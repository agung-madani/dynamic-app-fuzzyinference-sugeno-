```python
# define a function to return a list of unique elements from a given list
def unique(list1):
    unique_list = []
    for x in list1:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list
```

## Create Membership Functions for Input Variables

```python
# initialize empty list for storing input variables
him_var_in = []

# define some lists for fuzzy membership functions
ytrimf = [0,1,0]
ytrapmf1 = [1,1,1,0]
ytrapmf2 = [0,1,1,1]
ytrapmf3 = [0,1,1,0]
xtrimf = ['p','q','r']
xtrapmf = ['p','q','r','s']
```

For example, I want to make an application regarding the selection of students who are eligible to receive scholarships.

The variables input i use is:
1. Students' GPA
2. Parents' Salaries

and the variable output is:
1. Eligibility Value

```python
# loop until user provides a valid integer for number of input variables
while True:
    try:
        # ask user for number of input variables
        jum_var_in = int(input("Enter how many input variables: "))
    except ValueError:
        # handle invalid input (non-integer)
        print("Do not enter a value other than an integer")
        continue
    if jum_var_in < 1:
        # handle invalid input (number of variables less than 1)
        print('Enter at least 1')
        continue
    else:
        # exit loop with valid input
        break

# print number of input variables        
print(jum_var_in)
```

```python
# initialize empty list for storing input variables
var_in = []

# loop through number of input variables
for i in range(jum_var_in):
    # initialize empty list for storing current input variable information
    x = []
    # loop until user provides a valid input variable name
    while True:
        try:
            # ask user for input variable name
            a = input(f'Enter the name of the input variable {i+1}: ')
        except:
            # handle invalid input
            continue
        # check if input variable name has already been used
        for j in var_in:
            if a in j:
                print('Variable names cannot be the same')
                break
        else:
            # exit loop with valid input
            break
    # add input variable name to current input variable list
    x.append(a)
    # loop until user provides a valid number of membership functions for the input variable
    while True:
        try:
            # ask user for number of membership functions for input variable
            b = int(input(f'Enter how many sets of variables there are {a}: '))
        except ValueError:
            # handle invalid input (non-integer)
            print('Do not enter a value other than an integer')
            continue
        if b < 1:
            # handle invalid input (number of membership functions less than 1)
            print("Enter at least 1")
            continue
        else:
            # exit loop with valid input
            break
    # add number of membership functions to current input variable list
    x.append(b)
    # add current input variable list to list of all input variables
    var_in.append(x)

# print list of all input variables
print(var_in)
```

```python
# loop through each input variable
for i in var_in:
    # initialize empty list for storing membership function information for current input variable
    x = []
    # loop through number of membership functions for current input variable
    for j in range(i[1]):
        # initialize empty list for storing current membership function information
        y = []
        # loop until user provides a valid membership function name
        while True:
            try:
                # ask user for membership function name
                a = input(f'Enter the name of the set of variables {i[0]} to {j+1}: ')
            except:
                # handle invalid input
                continue
            # check if membership function name has already been used
            for k in x:
                if a in k:
                    print('Variable set names cannot be the same')
                    break
            else:
                # exit loop with valid input
                break
        # add membership function name to current membership function list
        y.append(a)
        # loop until user provides a valid membership function type (1 for TRIMF, 2 for TRAPMF)
        while True:
            try:
                # ask user for membership function type
                b = int(input(f'{a}, enter numbers 1(TRIMF), 2(TRAPMF): '))
            except ValueError:
                # handle invalid input (non-integer)
                print('Do not enter a value other than an integer')
                continue
            if b != 1 and b != 2:
                # handle invalid input (membership function type is not 1 or 2)
                print('Enter the correct number 1 or 2')
                continue
            else:
                # exit loop with valid input
                break
        # add membership function type to current membership function list
        y.append(b)
        # if membership function is TRIMF
        if b == 1:
            # ask user for domain parameters for TRIMF membership function
            print(f'Enter the TRIMF[p,q,r] parameter domain for {a}: ')
            # initialize empty list for storing domain parameters
            z = []
            # loop through domain parameters for TRIMF membership function
            for k in xtrimf:
                # loop until user provides a valid domain parameter value
                while True:
                    try:
                        # ask user for domain parameter value
                        while True:
                            try:
                                c = float(input(f'Enter value {k}: '))
                            except ValueError:
                                # handle invalid input (non-float)
                                print('Do not enter a value other than an float')
                            else:
                                # exit inner loop with valid input
                                break
                    except:
                        # handle invalid input
                        continue
                    # add domain parameter value to list of domain parameters
                    z.append(c)
                    # check if domain parameters are in increasing order
                    if z != sorted(z):
                        print('unordered domains')
                        # reset list of domain parameters
                        z = []
                        # continue outer loop to ask for domain parameter values again
                        continue
                    else:
                        # exit inner loop with valid input
                        break
            # set plot title and axis labels
            plt.title("Membership Function")
            plt.xlabel(i[0])
            plt.ylabel("Membership Degree")
            # plot TRIMF membership function
            plt.plot(z,ytrimf,label=a)
            # place legend outside plot area
            plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
            # display plot
            plt.show()
        # if membership function is TRAPMF
        if b == 2:
            # ask user for domain parameters for TRAPMF membership function
            print(f'Enter the domain parameter TRAPMF[p,q,r,s] for {a}: ')
            # initialize empty list for storing domain parameters
            z = []
            # loop until user provides valid domain parameter values for TRAPMF membership function
            while True:
                try:
                    # loop through domain parameters for TRAPMF membership function
                    for k in xtrapmf:
                        # loop until user provides a valid domain parameter value
                        while True:
                            try:
                                # ask user for domain parameter value
                                c = float(input(f'Enter value {k}: '))
                            except ValueError:
                                # handle invalid input (non-integer)
                                print('Do not enter a value other than an float')
                            else:
                                # exit inner loop with valid input
                                break
                        # add domain parameter value to list of domain parameters
                        z.append(c)
                except:
                    # handle invalid input
                    continue
                # check if domain parameters are in increasing order
                if z != sorted(z):
                    print('unordered domains')
                    # reset list of domain parameters
                    z = []
                    # continue outer loop to ask for domain parameter values again
                    continue
                else:
                    # exit outer loop with valid input
                    break
            # plot appropriate TRAPMF membership function based on domain parameters
            if z[0] == z[1]:
                # set plot title and axis labels
                plt.title("Membership Function")
                plt.xlabel(i[0])
                plt.ylabel("Membership Degree")
                # plot TRAPMF membership function
                plt.plot(z,ytrapmf1,label=a)
                # place legend outside plot area
                plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
                plt.show()
            elif z[2] == z[3]:
                # set plot title and axis labels
                plt.title("Membership Function")
                plt.xlabel(i[0])
                plt.ylabel("Membership Degree")
                # plot TRAPMF membership function
                plt.plot(z,ytrapmf2,label=a)
                # place legend outside
                plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
                # display plot
                plt.show()
            else:
                # set plot title and axis labels
                plt.title("Membership Function")
                plt.xlabel(i[0])
                plt.ylabel("Membership Degree")
                # plot TRAPMF membership function
                plt.plot(z,ytrapmf3,label=a)
                # place legend outside plot area
                plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
                # display plot
                plt.show()
        # add domain parameters to current membership function list
        y.append(z)
        # add current membership function list to list of membership functions for current input variable
        x.append(y)
    # add list of membership functions for current input variable to list of all membership functions
    him_var_in.append(x)

# print list of all membership functions
print(him_var_in)
```

```python
# set plot title
plt.title("Membership Function")
# loop through each input variable and its associated membership functions
for count,i in enumerate(var_in):
    # set x-axis label as input variable name
    plt.xlabel(i[0])
    # set y-axis label
    plt.ylabel("Membership Degree")
    # loop through membership functions for current input variable
    for j in range(len(him_var_in[count])):
        # if membership function is TRIMF
        if him_var_in[count][j][1] == 1:
            # plot TRIMF membership function
            plt.plot(him_var_in[count][j][2],ytrimf,label=him_var_in[count][j][0])
        # if membership function is TRAPMF
        elif him_var_in[count][j][1] == 2:
            # plot appropriate TRAPMF membership function based on domain parameters
            if him_var_in[count][j][2][0] == him_var_in[count][j][2][1]:
                plt.plot(him_var_in[count][j][2],ytrapmf1,label=him_var_in[count][j][0])
            # if membership function is TRAPMF with domain parameters [p,q,r,s] such that p=q
            elif him_var_in[count][j][2][2] == him_var_in[count][j][2][3]:
                # plot TRAPMF membership function
                plt.plot(him_var_in[count][j][2],ytrapmf2,label=him_var_in[count][j][0])
            # if membership function is TRAPMF with domain parameters [p,q,r,s] such that p!=q and r!=s
            else:
                # plot TRAPMF membership function
                plt.plot(him_var_in[count][j][2],ytrapmf3,label=him_var_in[count][j][0])
    # place legend outside plot area
    plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
    # display plot
    plt.show()
```

## Create Rules

```python
# Initialize empty list to store the rules
rules = []

# Function to add a rule to the list
def add_rules():
    # Initialize empty list to store the variables in the current rule
    x = []
    # Iterate over the input variables
    for i in range(jum_var_in):
        # Get the value of the current input variable and add it to the list
        globals()["as" + str(i + 1)] = globals()["varin" + str(i + 1)].get()
        x.append(globals()["as" + str(i + 1)])
    # Get the value of the output variable and add it to the list
    x.append(varout.get())
    # If the current rule is not already in the list, add it
    if x not in rules:
        rules.append(x)
    # Update the displayed list of rules
    show_rules()

# Function to display the list of rules
def show_rules():
    # If there are rules in the list
    if len(rules) > 0:
        # Iterate over the rules
        for i in range(len(rules)):
            # Initialize empty list to store the conditions for the current rule
            qw = []
            # If there is more than one input variable in the current rule
            if len(rules[i]) > 2:
                # Iterate over the input variables in the current rule
                for j in range(len(rules[i])-2):
                    # Create a string to represent the current condition and add it to the list
                    bs = f'AND {var_in[j+1][0]} {rules[i][j+1]}'
                    qw.append(bs)
            # Create a label to display the current rule
            globals()["showrule" + str(i)] = Label(root, text=f'IF {var_in[0][0]} {rules[i][0]} {qw} THEN {var_out[0]} {rules[i][len(rules[i])-1]}')
            # Display the label
            globals()["showrule" + str(i)].grid(row=i+5, column=1, columnspan=2)
            # Create a label to display the index of the current rule
            globals()["delrule" + str(i)] = Label(root, text=f'{i+1}')
            # Display the label
            globals()["delrule" + str(i)].grid(row=i+5, column=3)

# Function to delete a rule from the list
def del_rules():
    # Get the index of the rule to delete
    variabel = int(e.get())
    # Iterate over the rules
    for i in range(len(rules)):
        # Destroy the label for the current rule
        globals()["showrule" + str(i)].destroy()
        # Destroy the label for the index of the current rule
        globals()["delrule" + str(i)].destroy()
    # Delete the specified rule from the list
    del rules[variabel-1]
    # Update the displayed list of rules
    show_rules()
```

```python
# Create the main window
root = Tk()
# Set the title of the window
root.title("Rules")

# Create a label for the first input variable
varLabel1 = Label(root, text=f'IF {var_in[0][0]}')
# Display the label
varLabel1.grid(row=1, column=1)

# Iterate over the remaining input variables
for i in range(jum_var_in-1):
    # Create a label for the current input variable
    globals()["varLabel" + str(i + 2)] = Label(root, text=f'AND {var_in[i+1][0]}')
    # Display the label
    globals()["varLabel" + str(i + 2)].grid(row=1, column=i+2)    

# Iterate over the input variables
for i in range(jum_var_in):
    # Create a dropdown menu for the current input variable
    globals()["varin" + str(i + 1)] = Combobox(root, width = 27, textvariable = StringVar())
    # Set the options for the dropdown menu to be the sets of input variables for the current input variable
    globals()["varin" + str(i + 1)]['values'] = choices_in[i]
    # Display the dropdown menu
    globals()["varin" + str(i + 1)].grid(row=2, column=i+1)
    # Set the default value for the dropdown menu to be the first option
    globals()["varin" + str(i + 1)].current(0)

# Create a label for the output variable
varLabel0 = Label(root, text=f'THEN {var_out[0]}')
# Display the label
varLabel0.grid(row=3, column=1)

# Create a dropdown menu for the output variable
varout = Combobox(root, width = 27, textvariable = StringVar())
# Set the options for the dropdown menu to be the sets of output variables
varout['values'] = choices_out
# Display the dropdown menu
varout.grid(row=4, column=1)
# Set the default value for the dropdown menu to be the first option
varout.current(0)

# Create a text field for inputting a rule number to delete
e = Entry(root, width=27, borderwidth=5)
# Display the text field
e.grid(row=5, column=4, ipadx = 5)
# Create a label for the delete rule text field
qlabel = Label(root, text="Delete Rule Number", width=27, borderwidth=5)
# Display the label
qlabel.grid(row=4, column=4, ipadx = 5)

# Create a button for adding a rule
myButton = Button(root, text="Add", command=add_rules)
# Display the button
myButton.grid(row=4, column=2, ipadx = 27)

# Create a button for displaying the current rules
myButton1 = Button(root, text="Rules", command=show_rules)
# Display the button
myButton1.grid(row=4, column=3, ipadx = 27)

# Create a button for deleting a rule
myButton2 = Button(root, text="Delete", command=del_rules)
# Display the button
myButton2.grid(row=6, column=4, ipadx = 27)

# Run the main loop for the window
root.mainloop()
```

## Sugeno Method Fuzzy Inference Application GUI

```python
def inferensi_fuzzy():
    # Initialize variables "nilaiinput1" through "nilaiinput(jum_var_in)"
    # using the global namespace and the loop index
    for i in range(jum_var_in):
        globals()["nilaiinput" + str(i + 1)] = int(globals()["scalein" + str(i + 1)].get())
    
    # Initialize empty list to store values for him_var_input
    him_var_input = []
    # Loop through choices_in and domain_in, and append elements that meet certain conditions
    # to him_var_input
    for count, (j,k) in enumerate(zip(choices_in,domain_in)):
        temp = []
        for l in range(len(j)):
            if k[l][0] <= globals()["nilaiinput" + str(count + 1)] <= k[l][-1]:
                temp.append(j[l])
        him_var_input.append(temp)
    
    # Initialize empty list to store values for him_y_input
    him_y_input = []
    # Loop through choices_in and domain_in, and append elements that meet certain conditions
    # to him_y_input
    for count, (j,k) in enumerate(zip(choices_in,domain_in)):
        temp = []
        for l in range(len(k)):
            if any(j[l] in s for s in him_var_input[count]):
                if len(k[l]) == 4:
                    if k[l][0] == k[l][1]:
                        ytemp = max(min(1,(k[l][3]-globals()["nilaiinput" + str(count + 1)])/(k[l][3]-k[l][2])),0)
                    elif k[l][2] == k[l][3]:
                        ytemp = max(min((globals()["nilaiinput" + str(count + 1)]-k[l][0])/(k[l][1]-k[l][0]),1),0)
                    else:
                        ytemp = max(min((globals()["nilaiinput" + str(count + 1)]-k[l][0])/(k[l][1]-k[l][0]),1,(k[l][3]-globals()["nilaiinput" + str(count + 1)])/(k[l][3]-k[l][2])),0)
                elif len(k[l]) == 3:
                    ytemp = max(min((globals()["nilaiinput" + str(count + 1)]-k[l][0])/(k[l][1]-k[l][0]),(k[l][2]-globals()["nilaiinput" + str(count + 1)])/(k[l][2]-k[l][1])),0)
                temp.append(ytemp)
        him_y_input.append(temp)
    
    # Initialize empty list to store values for him_var_output
    him_var_output = []
    # Loop through the list "rules" and append elements to him_var_output if they meet certain conditions
    for i in rules:
        count = 0
        while count < jum_var_in:
            if i[count] in him_var_input[count]:
                if count == (jum_var_in-1):
                    him_var_output.append(i[-1])
                count += 1
            else:
                break
    
    # Initialize empty list to store values for him_y_output
    him_y_output = []
    # Generate a list of combinations using itertools.product and the elements of him_y_input
    # Calculate the maximum of each combination and append it to him_y_output
    for combination in itertools.product(*him_y_input):
        minimum = min(combination)
        him_y_output.append(minimum)
        
    # Calculate the unique elements of him_var_output
    him_var_output_set = unique(him_var_output)
    # Initialize empty list to store values for him_y_output_set
    him_y_output_set = []
    # Calculate the maximum of the corresponding elements in him_y_output for each element in him_var_output_set
    # and append it to him_y_output_set
    for i in him_var_output_set:
        ytemp = []
        for j,k in zip(him_var_output,him_y_output):
            if i == j:
                ytemp.append(k)
        him_y_output_set.append(max(ytemp))
    
    # Initialize empty list to store values for him_x_output_set
    him_x_output_set = []
    # Calculate the middle element of the domain of each unique element in him_var_output_set
    # and append it to him_x_output_set
    for i in him_var_output_set:
        for j,k in zip(choices_out,domain_out):
            if i == j:
                if len(k) == 3:
                    him_x_output_set.append(k[1])
                elif len(k) == 4:
                    if k[0] == k[1]:
                        him_x_output_set.append(k[2])
                    elif k[2] == k[3]:
                        him_x_output_set.append(k[1])
                    else:
                        him_x_output_set.append(k[1])
                        him_x_output_set.append(k[2])
    
    # Calculate the defuzzified output by summing the products of each element in him_var_output_set, him_y_output_set,
    # and him_x_output_set and dividing by the sum of him_y_output_set
    defuzzified = sum(i*j for i,j in zip(him_y_output_set,him_x_output_set))/sum(him_y_output_set)
    
    # Update the output label with the defuzzified output
    output.config(text=defuzzified)
```
